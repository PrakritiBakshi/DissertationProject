```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import random
import time
```

## collect the page and make a soup

```python
my_headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36"}

url = 'https://www.amazon.in/s?k=face+lightening+cream&page=2&crid=2ZE7HHTR6GLBE&qid=1659673255&sprefix=face+lightening+cream%2Caps%2C77&ref=sr_pg_1'

req = requests.get(url, headers=my_headers)
```

```python
soup =  bs(req.content)

# empty list for the dictionaries
list_of_dicts = []
```

## collect all product links and make list

```python
# Fetch product links as list of Tag Objects
links = soup.find_all("a", attrs={'class':'a-link-normal s-no-outline'})

# Store the links
links_list = []

# Loop for extracting links from Tag Objects
for link in links:
    links_list.append(link.get('href'))
    
```

## parsing for each product
        name, price, benefit, uses, about / description

```python
#introduce delays for multiple requests

sleep_times = [120,60,30]

#limit number of queries while testing to not get blocked by amazon

for idx, link in enumerate(links_list):
    #if idx > 0 and idx < 5:
         #links_list.append(link.get('href'))
        my_headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36"}

        url = "https://www.amazon.in" + link
        print(url)

        req = requests.get(url, headers=my_headers)

        print(req.status_code)
        
        sleep_time = random.choice(sleep_times)
        
        if req.status_code ==200:
        
            soup =  bs(req.content)


            # name
            for a in soup.find_all('span', class_='a-size-large product-title-word-break'):
                prod = a.get_text(strip=True).strip('\n')

            # price
            prix = soup.find_all('span', class_='a-price-whole')[0].get_text().strip('.')

            # benefit
            for x in soup.find_all('tr', class_="a-spacing-small po-product_benefit"):
                for y in x.find_all('td', class_ = 'a-span9'):
                    prodben = y.get_text(strip=True)

            # use
            for x in soup.find_all('tr', class_="a-spacing-small po-specific_uses_for_product"):
                for y in x.find_all('td', class_ = 'a-span9'):
                    produse = y.get_text(strip=True)

            # about
            join_list=[]
            for x in soup.find_all('div', class_='a-section a-spacing-medium a-spacing-top-small'):
                for y in x.find_all('span', class_ = 'a-list-item'):
                    productAbout = y.get_text(strip=True)
                    join_list.append(productAbout)
                full_str = ' '.join(join_list) 

            # dictionary for this product
            list_of_dicts.append({"product": prod, "price": prix, "benefits":prodben, "use": produse, "description":full_str, "url":url})
            #print(list_of_dicts)
            
        if idx<len(links_list)-1:
            print('Waiting for {} minutes for next hit'.format(sleep_time/60))
            time.sleep(sleep_time)
```

## make a dataframe and download as a csv

```python
df = pd.DataFrame(list_of_dicts)
df
```
